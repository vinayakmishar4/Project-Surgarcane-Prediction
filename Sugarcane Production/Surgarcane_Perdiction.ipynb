# SUGARCANE PREDICTION

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("List of Countries by Sugarcane Production.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Production (Tons)"] = df["Production (Tons)"].str.replace(".","")
df["Production per Person (Kg)"] = df["Production per Person (Kg)"].str.replace(".","").str.replace(",",".")
df["Acreage (Hectare)"] = df["Acreage (Hectare)"].str.replace(".","")
df["Yield (Kg / Hectare)"]= df["Yield (Kg / Hectare)"].str.replace(".","").str.replace(",",".")
```

```python
df.head()
```

```python
df.info()
```

```python
df.rename(columns={"Unnamed: 0" : "Index"},inplace=True)
df.rename(columns= {"Production (Tons)": "Production(Tons)"}, inplace = True)
df.rename(columns= {"Production per Person (Kg)": "Production_per_person(Kg)"}, inplace = True)
df.rename(columns= {"Acreage (Hectare)": "Acreage(Hectare)"}, inplace = True)
df.rename(columns= {"Yield (Kg / Hectare)": "Yield(Kg/Hectare)"}, inplace = True)
```

```python
df.head(1)
```

```python
df["Production(Tons)"] = df["Production(Tons)"].astype(float)
df["Production_per_person(Kg)"] = df["Production_per_person(Kg)"].astype(float)
df["Acreage(Hectare)"] = df["Acreage(Hectare)"].astype(float)
df["Yield(Kg/Hectare)"] = df["Yield(Kg/Hectare)"].astype(float)
```

```python
df.head(1)
```

```python
mean_Value=df['Acreage(Hectare)'].mean()
df['Acreage(Hectare)'].fillna(mean_Value,inplace=True)
```

```python
df.isnull().sum()
```

```python
mean_Value=df['Yield(Kg/Hectare)'].mean()
df['Yield(Kg/Hectare)'].fillna(mean_Value,inplace=True)
```

```python
df.isnull().sum()
```

# ANALYSIS FOR CONTINENT 

```python
df_continet=df.groupby('Continent').sum()
```

```python
df_continet
```

```python
df_continet['Production(Tons)'].sort_values(ascending=False).plot(kind='bar')
```

```python
df_continet['Acreage(Hectare)'].sort_values(ascending=True).plot(kind='bar')
```

```python
df_continet['number_of_Countries']=df.groupby(df['Continent']).count()['Country']
```

```python
df_continet
```

```python
contient_names=df_continet.index.to_list()
sns.lineplot(data=df_continet,x=df_continet['number_of_Countries'],y=df_continet['Production(Tons)'])
plt.xticks(df_continet['number_of_Countries'],contient_names,rotation=90)
plt.show()
```

```python
df_continet.corr(numeric_only=True)
```

```python
df_continet['Production(Tons)'].plot(kind='pie',autopct='%.2f')
```

Summary,That We Increse the Land can increse the Production in tons
